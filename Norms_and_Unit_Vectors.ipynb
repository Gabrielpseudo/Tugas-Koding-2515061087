```python
#Buat vector v dengan elemen (3,4)
v = np.array([[25,15,6,10 ],
              [10,6,15,25],
              [25,15,6,10]])
print("Vector v: ",v )
v.shape
```

```python
#Hitung nilai L2 Norm dari vector v
norm = np.linalg.norm(v)
print("nilai L2 norm: ", norm)
```

```python
# Menghitung unit vector
unit_v = v / norm
print("Unit vector:", unit_v)
```

```python
#Hitung panjang unit vector dari vector v
unit_v = np.linalg.norm(unit_v)
print("Panjang unit vektor: ", np.linalg.norm(unit_v))
```
